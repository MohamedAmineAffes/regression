```python
Train_Test_2018=df_2018.dropna()
print(Train_Test_2018.shape)
Train_Test_2018.head()
```

# ALGORITHM

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(Train_Test_2018.iloc[:,0:69],Train_Test_2018.iloc[:,69:70], test_size= 0.2, random_state=12)
```

```python
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import GridSearchCV
params = {'n_neighbors':[2,3,4,5,6,7,8,9]}

knn1 = neighbors.KNeighborsRegressor()

model2018 = GridSearchCV(knn1, params, cv=5)
model2018.fit(X_train,y_train)
model2018.best_params_
```

```python
params = {'n_neighbors':[3]}
model2018 = GridSearchCV(knn1, params, cv=5)
model2018.fit(X_train,y_train)
```

```python
score_knn1 = model2018.score(X_test, y_test)
score_knn1
```

```python
import math
# Predictions
preds2018 = model2018.predict(X_test)
```

# RESULTS FOR 2018

```python
for x in preds2018:
    print(math.ceil(x))
```
